```python
from openai import OpenAI

openai_api_key = userdata.get('OPENAI_API_KEY')

# 1. 初始化 Client (填入你的 OpenAI API Key)
client = OpenAI(
    api_key=openai_api_key
)

# 2. 呼叫模型生成內容
response = client.chat.completions.create(
    model="gpt-4o",  # 或使用 gpt-3.5-turbo
    messages=[
        {"role": "system", "content": "你是一位專業的軟體工程師。"},
        {"role": "user", "content": "哈囉，這是一封測試訊息，請用繁體中文回應。非常簡短地解釋一下 JavaScript 的 Hoisting。"}
    ]
)

# 3. 印出結果
print(response.choices[0].message.content)
```

# 3. 使用 Google Gemini 函式庫

## 3.1 Prepare

```python
# 安裝 Google Gen AI 函式庫

!pip install -U -q google-genai
```

```python
# genai 只能使用 Gemini AI KEY

from google import genai
from google.colab import userdata
```

```python
gemini_api_key = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=gemini_api_key)
```

## 3.2 Example Code

```python
# 使用 Gemini API

response = client.models.generate_content(
    model="gemini-2.0-flash", # 或 gemini-1.5-flash
    contents="哈囉，這是一封測試訊息"
)

print(response.text)
```

```python
# 列出 Gemini 所有可用的模型

for model in client.models.list():
    print(f"模型名稱: {model.name}")
    # 針對新版 SDK，使用 supported_actions
    # 某些模型可能沒有此屬性，使用 getattr 安全取值
    actions = getattr(model, 'supported_actions', '未提供')
    print(f"支援的功能: {actions}")
    print("-" * 20)
```

```python
# 列出消耗的時間
%%time

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="簡單三句話解釋焦土政策"
)

print(response.text)
```

```python
# 即時看到生成過程

response = client.models.generate_content_stream(
    model="gemini-2.5-flash",
    contents="簡單解釋 design pattern 中的 singleton"
)

for chunk in response:
    print(chunk.text)
    print("-" * 80)
```

## 3.3 Example Code: 與大模型即時通訊對話

### 3.3.1 一般對話

```python
# 建立對話會話，history 可以預設為空列表

chat = client.chats.create(model="gemini-2.5-flash", history=[])

# 傳送訊息
response = chat.send_message("以一句話向小孩解釋電腦如何運作。")

# 直接存取 text 屬性
print(response.text)
```

### 3.3.2 串流對話 (send_message_stream)

```python
# 使用串流方法進行對話
response = chat.send_message_stream("那如果對高中生進行更詳細的解釋呢？")

for chunk in response:
    print(chunk.text)
    print("-" * 80)
```

### 3.3.3 觀察 history 內容

```python
print(chat.get_history())
```

### 3.3.4 美化 history 格式

```python
for entry in chat.get_history():
    # 取得角色名稱 (user 或 model)
    role_name = "【使用者】" if entry.role == "user" else "【AI 助手】"

    # 將該內容中所有的 parts 合併（有時一個 Content 裡有多個 Part）
    content_text = "".join([part.text for part in entry.parts if part.text])

    print(f"{role_name}:")
    print(content_text)
    print("-" * 30)
```

```python
import pandas as pd

data = []
for entry in chat.get_history():
    data.append({
        "Role": entry.role,
        "Text Snippet": "".join([part.text for part in entry.parts if part.text])[:50] + "..." # 只顯示前50字
    })

df = pd.DataFrame(data)
display(df)
```
